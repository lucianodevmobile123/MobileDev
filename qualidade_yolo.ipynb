```python
from google.colab import files
from PIL import Image as PILImage
import matplotlib.pyplot as plt

def carregar_imagem_local():
    print("📂 Faça o upload da imagem do solo (JPG ou PNG):")
    uploaded = files.upload()

    for nome_arquivo in uploaded.keys():
        print(f"✅ Imagem '{nome_arquivo}' carregada.")
        img = PILImage.open(nome_arquivo)
        plt.imshow(img)
        plt.axis('off')
        plt.title("Imagem carregada do solo")
        plt.show()
        return nome_arquivo

```

```python
import torch

import torch

def carregar_modelo_yolo(modelo_path='best.pt'):
    print(f"🚀 Carregando modelo YOLO personalizado a partir de '{modelo_path}'...")
    modelo = torch.hub.load('ultralytics/yolov5', 'custom', path=modelo_path, force_reload=True)
    print("✅ Modelo carregado com sucesso.")
    return modelo

```

```python
from IPython.display import Image as IPyImage, display
import os

def detectar_e_exibir(modelo, imagem_path):
    print("🔍 Executando detecção na imagem...")
    resultados = modelo(imagem_path)

    # Mostrar as detecções no console
    resultados.print()

    # Salvar imagem com as anotações
    resultados.save()

    # Achar a imagem anotada salva automaticamente no diretório padrão
    diretorio_resultado = resultados.save_dir
    imagem_anotada = os.path.join(diretorio_resultado, os.listdir(diretorio_resultado)[0])

    print("🖼️ Resultado da detecção:")
    display(IPyImage(filename=imagem_anotada))

```

```python
# 🔁 Pipeline completo

# 1. Upload da imagem
imagem_path = carregar_imagem_local()

# 2. Upload do modelo YOLOv5 (.pt treinado)
print("📦 Agora, envie seu modelo YOLOv5 (arquivo .pt)")
modelo_file = files.upload()
modelo_nome = list(modelo_file.keys())[0]

# 3. Carregamento do modelo
modelo_yolo = carregar_modelo_yolo(modelo_nome)

# 4. Detecção
detectar_e_exibir(modelo_yolo, imagem_path)
```
