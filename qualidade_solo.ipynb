```python
# Bibliotecas necessarias
from google.colab import files
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
%pip install ultralytics
```

```python
def carregar_imagem_solo():
    print("üåæ Por favor, selecione uma imagem do solo para an√°lise:")
    uploaded = files.upload()  # Abre o seletor de arquivos
    for nome_arquivo in uploaded.keys():
        imagem = Image.open(nome_arquivo)
        print(f"‚úÖ Imagem '{nome_arquivo}' carregada com sucesso.")
        return imagem
```

```python
def exibir_imagem(imagem):
    plt.figure(figsize=(6, 4))
    plt.imshow(imagem)
    plt.axis('off')
    plt.title("üñºÔ∏è Imagem do solo selecionada")
    plt.show()
```

```python
def analisar_qualidade_solo(imagem):
    # Reduz o tamanho para acelerar o processamento e normaliza
    imagem_redimensionada = imagem.resize((100, 100))
    matriz = np.array(imagem_redimensionada) / 255.0

    # Calcula a m√©dia de cor
    media_rgb = matriz.mean(axis=(0, 1))

    print(f"üîç M√©dia das cores - Vermelho: {media_rgb[0]:.2f}, Verde: {media_rgb[1]:.2f}, Azul: {media_rgb[2]:.2f}")

    # L√≥gica simulada para avalia√ß√£o do solo
    if media_rgb[0] < 0.3 and media_rgb[1] < 0.3:
        return "‚ùå Solo com baixa fertilidade ‚Äî Pouco indicado para o cultivo de culturas tradicionais."
    elif media_rgb[0] > 0.5 and media_rgb[1] > 0.4:
        return "‚úÖ Solo f√©rtil ‚Äî Adequado para o plantio de milho, feij√£o e mandioca."
    else:
        return "‚ö†Ô∏è Solo mediano ‚Äî Recomendado realizar corre√ß√£o com fertilizantes antes do plantio."
```

```python
imagem_solo = carregar_imagem_solo()
exibir_imagem(imagem_solo)

resultado_analise = analisar_qualidade_solo(imagem_solo)
print("\nüå± Resultado da an√°lise do solo:")
print(resultado_analise)
```
