```python
# Bibliotecas necessarias
from google.colab import files
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
%pip install ultralytics
```

```python
def carregar_imagem_solo():
    print("🌾 Por favor, selecione uma imagem do solo para análise:")
    uploaded = files.upload()  # Abre o seletor de arquivos
    for nome_arquivo in uploaded.keys():
        imagem = Image.open(nome_arquivo)
        print(f"✅ Imagem '{nome_arquivo}' carregada com sucesso.")
        return imagem
```

```python
def exibir_imagem(imagem):
    plt.figure(figsize=(6, 4))
    plt.imshow(imagem)
    plt.axis('off')
    plt.title("🖼️ Imagem do solo selecionada")
    plt.show()
```

```python
def analisar_qualidade_solo(imagem):
    # Reduz o tamanho para acelerar o processamento e normaliza
    imagem_redimensionada = imagem.resize((100, 100))
    matriz = np.array(imagem_redimensionada) / 255.0

    # Calcula a média de cor
    media_rgb = matriz.mean(axis=(0, 1))

    print(f"🔍 Média das cores - Vermelho: {media_rgb[0]:.2f}, Verde: {media_rgb[1]:.2f}, Azul: {media_rgb[2]:.2f}")

    # Lógica simulada para avaliação do solo
    if media_rgb[0] < 0.3 and media_rgb[1] < 0.3:
        return "❌ Solo com baixa fertilidade — Pouco indicado para o cultivo de culturas tradicionais."
    elif media_rgb[0] > 0.5 and media_rgb[1] > 0.4:
        return "✅ Solo fértil — Adequado para o plantio de milho, feijão e mandioca."
    else:
        return "⚠️ Solo mediano — Recomendado realizar correção com fertilizantes antes do plantio."
```

```python
imagem_solo = carregar_imagem_solo()
exibir_imagem(imagem_solo)

resultado_analise = analisar_qualidade_solo(imagem_solo)
print("\n🌱 Resultado da análise do solo:")
print(resultado_analise)
```
