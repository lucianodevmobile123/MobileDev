```python
from ultralytics import YOLO
model = YOLO('yolov8n.pt')
```

```python
results = model('https://ultralytics.com/images/bus.jpg')
```

```python
for result in results:
  boxes = result.boxes
  result.save(filename='result.jpg')
```
